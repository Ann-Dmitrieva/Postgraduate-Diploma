```python
import math
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
import pandas as pd
from scipy.optimize import curve_fit
from sklearn.metrics import r2_score
import seaborn as sns
```

Теплофизические свойства при Т = 100С

```python
Cp = 4216 #Дж/кг*К - вода
r = 2257200 #Дж/кг
ρ_w = 958.1 #м3/кг - вода
ρ_v = 0.597 #м3/кг - пар
η = 2790 / (10 ** 7) # Н*c/м2 - динамическая 
λ = 0.68 #Вт/м К
g = 9.8
ν = η / ρ_w #м2/с - кинематическая
σ = 0.05891 #Н/м - поверхностное натяжение
a_t = λ / (Cp * ρ_w)
ΔT = 7.49
a_t
```

```python
Ja = (Cp * ΔT * ρ_w) / (r * ρ_v)
Ja0 = (Cp * ρ_w) / (r * ρ_v)
Pr = (η * Cp) / λ
Ar = (g / (ν ** 2)) * ((σ / (g * ρ_w)) ** (3 / 2))
K = ((Ja / Pr) ** 2) / Ar
Ja0
```

```python
d0 = 0.148 * math.sqrt(1 + 100000 * K) * math.sqrt(σ / (g * ρ_w))
d0_b = 0.148 * math.sqrt(1 + 100000 * K)
```

Обработка экспериментальных данных

```python
tests = pd.read_csv('16.06.2023 Water d=1mm.csv', sep=';')
tests = tests.drop(index=[14, 15, 16, 17, 18])
tests
```

```python
tests = tests.set_index('Замер')
tests = tests.replace(',', '.', regex=True)
tests = tests.astype(float)
```

```python
tests.columns = ['U, V', 'I, A', 'P, W', 'q, W/m2']
tests['ΔT'] = (tests['q, W/m2'] ** 0.3) / 3
tests
```

```python
tests['D0, m'] = [1207.5, 1162.2, 1192.8, 1238.7, 1303.7, 1396.5, 1537, 1671.4, 1747.5, 1879, 2207.9, 2546.8, 2654.6, 2419.7]
tests
```

```python
tests = tests.rename(columns={'D0, m' : 'D0, mkm'})
```

```python
x = np.linspace(7, 13, 5)
graph, axes = plt.subplots()
axes.errorbar(tests['ΔT'], tests['D0, mkm'], yerr=70, fmt='o', capsize=3, markersize=2)
axes.plot(x, 0.25 * 1000000 * np.sqrt(1 + 100000 * (((((Cp * x * ρ_w) / (r * ρ_v)) / Pr) ** 2) / Ar)) * math.sqrt(σ / (g * ρ_w)))
```

Подбираем а:

```python
def func(x, a):
    return a * 1000000 * np.sqrt(1 + 100000 * (((((Cp * x * ρ_w) / (r * ρ_v)) / Pr) ** 2) / Ar)) * math.sqrt(σ / (g * ρ_w))
popt, pcov = curve_fit(func, tests['ΔT'], tests['D0, mkm'])
a_opt = popt
D0_fit = func(tests['ΔT'], *popt)
D0_fit = func(tests['ΔT'], a_opt)
r2 = r2_score(tests['D0, mkm'], D0_fit)
r2, a_opt
```

Скорость роста

```python
growth = pd.read_csv('bubble growth.csv', sep=';')
```

```python
growth.columns = ['t1', 'd1', 't2', 'd2', 't5', 'd5', 't6', 'd6', 't7', 'd7', 't10', 'd10']
```

```python
growth = growth.drop(index=0)
```

```python
growth1 = growth[['t1', 'd1']]
growth2 = growth[['t2', 'd2']]
growth5 = growth[['t5', 'd5']]
growth6 = growth[['t6', 'd6']]
growth7 = growth[['t7', 'd7']]
growth10 = growth[['t10', 'd10']]
```

```python
growth1 = growth1.dropna()
growth2 = growth2.dropna()
growth5 = growth5.dropna()
growth6 = growth6.dropna()
growth7 = growth7.dropna()
growth10 = growth10.dropna()
```

```python
growth_data = [
    {'t': growth1['t1'], 'd': growth1['d1']},
    {'t': growth2['t2'], 'd': growth2['d2']},
    {'t': growth5['t5'], 'd': growth5['d5']},
    {'t': growth6['t6'], 'd': growth6['d6']},
    {'t': growth7['t7'], 'd': growth7['d7']},
    {'t': growth10['t10'], 'd': growth10['d10']}
]
for i in growth_data:
    i['t'] = i['t'].astype('int32')
    i['d'] = i['d'].astype('int32')
    
temperatures = [7.49, 7.77, 8.7, 9.41, 10.03, 11.4]
def growth_func(x, temp):
    gamma = 0.3
    beta = 6
    return 2000000 * np.sqrt(a_t * x / 1000000) * (
        gamma * ((Cp * temp * ρ_w) / (r * ρ_v)) + 
        np.sqrt((gamma ** 2) * (((Cp * temp * ρ_w) / (r * ρ_v)) ** 2) + 
                2 * beta * ((Cp * temp * ρ_w) / (r * ρ_v))))
fig, axes = plt.subplots(nrows=3, ncols=2, figsize=[12, 12])
axes = axes.flatten()
titles = ['ΔT = 7.49 K', 'ΔT = 7.77 K', 'ΔT = 8.7 K', 'ΔT = 9.41 K', 'ΔT = 10.03 K', 'ΔT = 11.4 K']

for ax, data, tempr, tit in zip(axes, growth_data, temperatures, titles):
    ax.errorbar(data['t'], data['d'],yerr=59, fmt='o', capsize=3, markersize=2)
    x_set = np.linspace(5, 5000, 100) 
    y_set = growth_func(x_set, tempr)
    ax.set_xlabel('t, mks')
    ax.set_ylabel('D, mkm')
    ax.set_title(tit)
    ax.axis([0, 5000, 0, 3000])
    ax.plot(x_set, y_set)
    ax.set_xticks(ax.get_xticks()[::])
    ax.set_yticks(ax.get_yticks()[::2])

plt.tight_layout()
plt.savefig('Water d=1mm growth velocity0.png')
```

```python
growth1 = growth1.astype('int32')
growth2 = growth2.astype('int32')
growth5 = growth5.astype('int32')
growth6 = growth6.astype('int32')
growth7 = growth7.astype('int32')
growth10 = growth10.astype('int32')
```

Рассчитываем коэффициенты детерминации для скорости роста:

```python
def func(x, temper):
    return 2000000 * np.sqrt(a_t * x / 1000000) * (0.3 * ((Cp * temper * ρ_w) / (r * ρ_v)) + 
                                 np.sqrt(0.09 * (((Cp * temper * ρ_w) / (r * ρ_v)) ** 2) + 
                                          12 * ((Cp * temper * ρ_w) / (r * ρ_v))))
temperatures = [7.49, 7.77, 8.7, 9.41, 10.03, 11.4]
growth_data = [
    {'t': growth1['t1'], 'd': growth1['d1']},
    {'t': growth2['t2'], 'd': growth2['d2']},
    {'t': growth5['t5'], 'd': growth5['d5']},
    {'t': growth6['t6'], 'd': growth6['d6']},
    {'t': growth7['t7'], 'd': growth7['d7']},
    {'t': growth10['t10'], 'd': growth10['d10']}]
for gd, tem in zip(growth_data, temperatures):
    D_fit = func(gd['t'], tem)
    r2 = r2_score(gd['d'], D_fit)
    print(r2)
```

Подбор общих гаммы и беты:

```python
growth_data = [
    {'t': growth1['t1'], 'd': growth1['d1'], 'temperature' : 7.49},
    {'t': growth2['t2'], 'd': growth2['d2'], 'temperature' : 7.77},
    {'t': growth5['t5'], 'd': growth5['d5'], 'temperature' : 8.7},
    {'t': growth6['t6'], 'd': growth6['d6'], 'temperature' : 9.41},
    {'t': growth7['t7'], 'd': growth7['d7'], 'temperature' : 10.03}]

all_data = pd.concat([pd.DataFrame(data) for data in growth_data])
```

```python
def combined_func(t, gamma, beta):
    temp = all_data['temperature']
    return 2000000 * np.sqrt(a_t * t / 1000000) * (
        gamma * ((Cp * temp * ρ_w) / (r * ρ_v)) + 
        np.sqrt((gamma ** 2) * (((Cp * temp * ρ_w) / (r * ρ_v)) ** 2) + 
                2 * beta * ((Cp * temp * ρ_w) / (r * ρ_v))))
                    
popt, pcov = curve_fit(combined_func, all_data['t'], all_data['d'], bounds=([0.1, 5], [0.5, 10]))
d_fit = combined_func(all_data['t'], *popt)

R2 = r2_score(all_data['d'], d_fit)
print( R2, popt)
```
